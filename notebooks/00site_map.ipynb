```python
import os # Interoperable file paths
import pathlib

import geopandas as gpd
import holoviews as hv
import hvplot.pandas
import panel as pn
import pandas as pd

import bokeh
from bokeh.resources import INLINE

import selenium # Dependency to save hvplot as png file
import phantomjs # Dependency to save hvplot as png file
```

```python
data_dir = os.path.join(
    '..',
    'data',
    'raw'
)
os.makedirs(data_dir, exist_ok=True)
```

```python
# National Fire Occurrence Dataset 1984-2024
boundary_url = ("https://edcintl.cr.usgs.gov/downloads/sciweb1/shared/"
                "MTBS_Fire/data/composite_data/burned_area_extent_shapefile/"
                "mtbs_perimeter_data.zip"

boundary_dir = os.path.join(data_dir, 'mtbs')
os.makedirs(boundary_dir, exist_ok=True)

boundary_path = os.path.join(boundary_dir, 'mtbs_bab.shp')

# Download once
if not os.path.exists(boundary_path):
    boundary_gdf = gpd.read_file(boundary_url)
    boundary_gdf = boundary_gdf[boundary_gdf.Incid_Name == 'CAMERON PEAK']
    boundary_gdf.to_file(boundary_path)

# Load from file
else:
    boundary_gdf = gpd.read_file(boundary_path)
```

```python
boundary_gdf
```

```python
cpfire_map = boundary_gdf.hvplot(
    geo=True, 
    tiles='EsriImagery',
    title=(f'{boundary_gdf.Incid_Name.values[0].title()} Wildfire'),
    alpha=0.4)

cpfire_map
```

```python
hv.extension('bokeh')
panel_object = pn.pane.HoloViews(cpfire_map)
pn.pane.HoloViews(cpfire_map).save('cpfire_map', embed=True, resources=INLINE)
```
